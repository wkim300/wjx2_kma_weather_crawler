```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
```

```python
conn = sqlite3.connect('../wdb.db', isolation_level=None)

# Read in 'cities' dataframe from wdb.db
cities = pd.read_sql_query('SELECT * FROM cities', conn)
conn.close()
```

```python
city_loc = cities[['lat', 'long']]
city_loc.plot.scatter(x='long', y='lat')
```
